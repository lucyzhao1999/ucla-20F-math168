```python
from networkx.generators.random_graphs import erdos_renyi_graph

n = 6
p = 0.5
g = erdos_renyi_graph(n, p)
print(g.nodes)
# [0, 1, 2, 3, 4, 5]
print(g.edges)
# [(0, 1), (0, 2), (0, 4), (1, 2), (1, 5), (3, 4), (4, 5)]
len(g.edges)
```

```python
n = 20
p = 0.4
numGraphs = 2000

numEdgesList = []
meanGraphDegreeList = []

for graphID in range(numGraphs):
    g = erdos_renyi_graph(n, p)
    numEdges = len(g.edges)
    meanDegree = numEdges* 2/n
    
    numEdgesList.append(numEdges)
    meanGraphDegreeList.append(meanDegree)
```

```python
import numpy as np 
np.mean(numEdgesList)
```

```python
np.mean(meanGraphDegreeList)
```
