```python
import networkx as nx 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
n=200
l=3
b=2*l
ppower = np.array([-2, -1.5, -1, -0.5, 0])
pList = [10**-2, 10**-1.5, 10**-1, 10**-0.5, 10**0]

cmax = 3*(b-2)/4/(b-1)
lmax = n/4/l

meanGeodDistList = []
globalClustCoeffList = []

for p in pList:
    G = nx.watts_strogatz_graph(n, b, p) 
    meanGeodDist = nx.average_shortest_path_length(G)
    globalClustCoeff = np.mean(list(nx.clustering(G).values()))
    
    meanGeodDistList.append(meanGeodDist)
    globalClustCoeffList.append(globalClustCoeff)

plt.plot(ppower, np.array(globalClustCoeffList)/cmax, label='clustering coeff')
plt.plot(ppower, np.array(meanGeodDistList)/lmax, label='mean geod distance')
plt.xlabel('p (in power of 10)')

plt.suptitle('N = {}, l = {}'.format(n, l))

plt.legend(framealpha=1, frameon=True);
plt.show()
```

```python
n=50
l=3
b=2*l
ppower = np.array([-2, -1.5, -1, -0.5, 0])
pList = [10**-2, 10**-1.5, 10**-1, 10**-0.5, 10**0]

cmax = 3*(b-2)/4/(b-1)
lmax = n/4/l

meanGeodDistList = []
globalClustCoeffList = []

for p in pList:
    a = []
    c = []
    for i in range(100):
        G = nx.watts_strogatz_graph(n, b, p) 
        meanGeodDist = nx.average_shortest_path_length(G)
        globalClustCoeff = np.mean(list(nx.clustering(G).values()))
        a.append(meanGeodDist)
        c.append(globalClustCoeff)
    
    meanGeodDistList.append(np.mean(a))
    globalClustCoeffList.append(np.mean(c))

plt.plot(ppower, np.array(globalClustCoeffList)/cmax, label='clustering coeff')
plt.plot(ppower, np.array(meanGeodDistList)/lmax, label='mean geod distance')
plt.xlabel('p (in power of 10)')
plt.suptitle('N = {}, l = {}'.format(n, l))

plt.legend(framealpha=1, frameon=True);
plt.show()
```
